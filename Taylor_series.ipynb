This code calculates an approximation of cos(x) using the Taylor series expansion up to a specified number of terms. The user inputs an angle in degrees and the number of terms 
ùëõ. The angle is converted to radians because Python‚Äôs trigonometric functions work in radians. The calculation starts with the first term of the cosine Taylor series (which is 1), then adds subsequent even-powered terms of the form $$(-1)^k \frac{x^{2k}}{(2k)!}$$ The loop iterates through even values (2, 4, 6, ‚Ä¶), and the variable sign alternates between +1 and ‚Äì1 to match the alternating nature of the series.

After computing the Taylor approximation, the code calculates the actual value of cos(x) using math.cos(). It then computes the absolute error and relative error to show how close the approximation is to the true value. Finally, the program prints the numerical approximation, the analytic value, and both error measurements. This code demonstrates how Taylor series can approximate trigonometric functions and how accuracy improves as more terms are included.

```python
import math
import numpy as np

# Input angle in degrees
x = int(input("x= "))
# Number of terms in the series
n = int(input("n= "))

# Convert to radians
z = math.radians(x)
# First term / Cosine approximation
cosine_approx = 1
# Alternating sign
sign = -1

# Taylor series calculation
# The loop iterates for i = 2, 4, 6, ..., up to the largest even number less than n+1
for i in range(2, (n+1), 2):
    numerator = z**i
    denominator = math.factorial(i)
    cosine_approx = cosine_approx + sign * (numerator / denominator)
    sign *= -1

# Analytic value of cos(x)
analytic_value = math.cos(z)

# Absolute and relative error
absolute_error = abs(cosine_approx - analytic_value)
relative_error = abs((cosine_approx - analytic_value) / analytic_value) * 100

# Output
print("cos x_numerical = {:.10f}".format(cosine_approx))
print("cos x_analytic = {:.10f}".format(analytic_value))
print("absolute_error = {}".format(absolute_error))
print("relative_error = {}%".format(relative_error))
```

This code computes an approximation of _sin(x)_ using the Taylor series expansion up to a specified number of terms. The user enters an angle in degrees and the number of series terms 
ùëõ. The angle is first converted into radians, since Python‚Äôs trigonometric functions operate in radians. The calculation begins with the first Taylor series term for sine, which is 
ùë•, and the program adds higher-order odd-powered terms of the form $$(-1)^k \frac{x^{2k+1}}{(2k+1)!}$$ The loop iterates through odd powers (3, 5, 7, ‚Ä¶), and the variable sign alternates between +1 and ‚Äì1 to match the alternating pattern in the sine series.

After computing the Taylor approximation, the code calculates the true value of sin(x) using math.sin(). It then determines the absolute and relative errors to measure how close the approximation is to the analytic value. Finally, the program prints the numerical approximation, the actual value, and both error metrics, illustrating how the Taylor series converges toward the true sine value as more terms are included.

```python
import math
import numpy as np

x = int(input("x= "))  # input angle in degrees
n = int(input("n= "))  # number of series terms

z = math.radians(x)  # convert to radians
s = z                # first term of sin(x) = x
sign = -1            # alternating sign

# Taylor series calculation for sine
for i in range(3, (n+1), 2):   # odd powers: 3, 5, 7, ...
    a = z**i
    b = math.factorial(i)
    s = s + sign * (a/b)
    sign *= -1

# analytic value of sin(x)
s_asli = math.sin(z)

# absolute and relative error
error_abs = abs(s - s_asli)
error_rel = abs((s - s_asli) / s_asli) * 100

# output
print("sin x_numerical = {:.10f}".format(s))
print("sin x_analytic = {:.10f}".format(s_asli))
print("absolute_error = {}".format(error_abs))
print("relative_error = {}%".format(error_rel))
```
